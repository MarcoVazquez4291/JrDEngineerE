```python
# After doing this I finally found the API with some of the information to answer the questions
# Unfortunately the API asks for the departure AND arrival airport, and me, only having MEX for the Benito Juarez airport had to think of something
# so I decided to google the most transited MEX-USA flights which is CDMX-LA and search for the airport code por LA(LAX)
url = "https://api.flightstats.com/flex/flightstatus/rest/v2/json/route/status/MEX/LAX/dep/2021/1/18?appId=9cef3402&appKey=+d753b69e68ba50f5598091b45c893d53&hourOfDay=0&utc=false&numHours=24&maxFlights=5"
#retrieved the correct URL from the interactive documentation and followed as previously done.
```

```python
r = requests.get(url)
```

```python
json_data = r.json()
```

```python
print(json.dumps(json_data, indent=2))
```

```python
for key, value in json_data.items():
    dep_airport = json_data["appendix"]["airports"][0]["name"]
    arr_airport = json_data["appendix"]["airports"][1]["name"]
    print(dep_airport, '-', arr_airport)
```

```python
for key, value in json_data.items():
    fstatus = json_data["flightStatuses"][0]["status"]
    print(fstatus)
```

```python
# So, now we can see the depparture airport and the arrival airport, 
# we can see that there were 3 flights(the API limited the info to 3 flights so it's not in the past 2 days lapse),
# and we can see that they were all delayed
```
